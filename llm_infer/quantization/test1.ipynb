```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_path = "G:/model_weights/models/Qwen/Qwen3-4B-Instruct-2507"
model = AutoModelForCausalLM.from_pretrained(model_path, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_path)
```

```python
torch.cuda.empty_cache()
```

```python
prompt = "简单介绍一下曹操"
messages = [
    {"role": "system", "content": "你是一位历史专家，尤其擅长三国史，对于东汉末年和三国时期的任务颇具见解。"},
    {"role": "user", "content": prompt}
]
```

```python
input = tokenizer.apply_chat_template(
    messages, tokenize=False, add_generation_prompt=True
)
# tokenizer.decode(input["input_ids"])
input
```

```python
input_ids = tokenizer([input], return_tensors="pt").to(model.device)
input_ids
```

```python
response = model.generate(**input_ids, max_new_tokens=512)
```

```python
response.shape
```

```python
res = [gen[len(pm):] for pm, gen in zip(input_ids["input_ids"], response)]
```

```python
res
```

```python
tokenizer.batch_decode(res, skip_special_tokens=True)
```
